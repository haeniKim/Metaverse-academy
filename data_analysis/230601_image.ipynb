```python
#실습

#1. 이미지 반으로 자르기
#2. 이미지 4등분해서 저장하기

#1
h, w = imgdata.shape[:2]
img_cut = imgdata[:h//2,:,:]
img_cut2 = imgdata[h//2:,:,:]
```

```python
Image.fromarray(img_cut2).save('winter_cut_b,jpg')
```
