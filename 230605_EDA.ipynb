### 1. 성별에 따른 급여차이

```python
#수익 결측치 데이터가 많으므로 삭제하기보다는 평균값으로 대체
sex_income = data.dropna(subset=['income']).groupby('sex')['income'].mean()
sex_income
```

```python
sex_income.plot.bar(rot=0)
```

```python
sex_income.loc['M'] / sex_income.loc['F']
```

```python
#독립표본 T검정
import scipy.stats as stats

sex_m = data[data['sex']=='M'].dropna()
sex_f = data[data['sex']=='F'].dropna()
```

```python
sex_m
```

```python
#남여간의 평균연봉은 약 1.8배 정도 차이가 있다.
stats.ttest_ind(sex_m['income'],sex_f['income'])
```

```python
# 초년: - 29중년:30-59, 노년: 60-
data = data.assign(agetype = np.where(data['age']<30,'초년',
                                      np.where(data['age']<60,'중년','노년')))
data
```

```python
data['agetype'].value_counts()
```

```python
data.groupby('agetype')['income'].mean()
```

```python
#각 age type별 성별의 수익 평균
agetype_sex_income = data.groupby(['agetype','sex'], as_index=False)['income'].mean()
agetype_sex_income
```

```python
sns.barplot(data=agetype_sex_income, x='agetype', y='income', hue='sex', order=['초년','중년','노년'])
```

```python
age_sex_income = data.dropna(subset=['income']).groupby(['sex','age'], as_index=False)['income'].mean()
```

```python
job_name = pd.read_excel('/content/drive/MyDrive/Metaverse Academy/data/(2021년 16차 한국복지패널조사) 조사설계서-가구용(beta1).xlsx', sheet_name='직종코드(2019 신분류)')
job_name
```

```python
job_name.rename(columns={'Unnamed: 3':'직업'}, inplace=True)
```

```python
job_name.rename(columns={'소분류':'jobcode', '직업':'job'}, inplace=True)
job_name
```

```python
job_data = data.merge(job_name, on='jobcode', how='left')
job_data
```

```python
#직업별로 그룹지어 평균 구하기

#직업, 수입이 없는 데이터 삭제
job_income_mean = job_data.dropna(subset=['job','income']).groupby('job', as_index=False)['income'].mean()
job_income_mean
```

```python
top10_income_job = job_income_mean.sort_values('income', ascending = False).head(10)
top10_income_job
```
