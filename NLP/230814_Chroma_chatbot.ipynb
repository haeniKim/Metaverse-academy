```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import random
import numpy as np
```

```python
#데이터를 구할 때마다 벡터로 만들어 csv로 저장하면 가져오기 편함 -> 서버 켜서 임베딩하면 오래걸림.
np.set_printoptions(precision=6, suppress=True) #지수로 나오지 않도록 소수점 자리수 처리
```

```python
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
model = model.to('cpu')
```

```python
embeding = model.encode(['안녕']) #문장을 벡터로 바꿔줌.
embeding.shape #768개의 특징을 저장
```

```python
df = pd.read_csv('ChatbotData.csv')
df
```

```python
import csv
with open('embeding.csv', 'w') as file:
  mapdata = []
  write = csv.writer(file)
  for temp in df['Q']:
    t = model.encode(temp)
    mapdata.append(t)

  write.writerows(mapdata)
```

```python
df1 = pd.read_csv('embeding.csv', header=None)
df1
```

```python
def chatbot_text(text):
  em_result = model.encode(text)

  co_result = []
  for temp in range(len(df1)):
    data = df1.iloc[temp]
    co_result.append(cosine_similarity([em_result], [data])[0][0])

  df['cos'] = co_result #유사도 정렬을 위해
  df_result = df.sort_values('cos', ascending=False)
  r = random.randint(0,5)
  return df_result.iloc[r]['A']
```

```python
chatbot_text('배고파')
```

```python
em_result = model.encode('안녕')
cosine_similarity([em_result], [df1.iloc[0]])[0][0]
```

```python
# 테이블 idx, embeding, metadata.
import chromadb

client = chromadb.Client()
collections = client.create_collection('chatbot') #테이블
```

```python
embeddings = []
metadata = []
ids = []

for temp in range(len(df1)):
  ids.append(str(temp+1)) #1부터 시작
  embeddings.append(df1.iloc[temp].tolist())
  metadata.append({'A': df.iloc[temp]['A']})
```

```python
df1.iloc[0].tolist() #리스트 타입으로 변환 필요
```

```python
collections.add(embeddings=embeddings, metadatas=metadata, ids = ids) #DB에 넣기
```

```python
chat_text = model.encode('점심 뭐먹지?')
query_result = collections.query(query_embeddings = [chat_text.tolist()], n_results=3)
query_result['metadatas']
```
