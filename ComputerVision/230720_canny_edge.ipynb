```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
src = cv2.imread('galaxy-soho.jpg')
dst = cv2.Canny(src, 50, 150) #edge값만 찾아서 보여줌.- 물체의 경우 외각선이 검출
cv2_imshow(dst)
```

```python
import numpy as np

src = cv2.imread('building.jpg')
edges = cv2.Canny(src, 50, 150)

lines = cv2.HoughLinesP(edges, 1.0, np.pi / 180. , 160, minLineLength=50, maxLineGap=5) #선을 그릴 수 있는 포인트 도출

dst = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)

for temp in range(lines.shape[0]):
  pt1 = (lines[temp][0][0], lines[temp][0][1])
  pt2 = (lines[temp][0][2], lines[temp][0][3])
  cv2.line(dst, pt1, pt2, (0, 0, 255), 2, cv2.LINE_AA)

cv2_imshow(dst)
```
